Data Normalization and Visualization



Explanation:

plt.scatter: This function is used to create a scatter plot.

subset['Student_ID']: Specifies the x-axis data, which in this case is the Student_ID. Each point in the scatter plot represents a student.

subset['Normalized_Score']: Specifies the y-axis data, which is the normalized score for each student.

color=colors.get(subject, 'black'): Chooses the color for each subject.
It looks up the color associated with the current subject in the colors dictionary. If the subject isn't found in the dictionary, it defaults to 'black'.

label=subject: Assigns a label to each subject, which is used in the legend to distinguish between different subjects in the plot.

This line is looped for each unique subject in the DataFrame, allowing the script to plot the normalized scores for all subjects on the same scatter plot, with different colors for each subject.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# 1. Load the CSV file into a pandas DataFrame
df = pd.read_csv('student_scores.csv')

# 2. Normalize the Score column (scaling between 0 and 1)
scaler = MinMaxScaler()
df['Normalized_Score'] = scaler.fit_transform(df[['Score']])

# 3. Create a scatter plot of Score for each Subject with different colors for each subject

# Define colors for each subject
colors = {'Digital Electronics': 'blue', 'Data Mining': 'green', 'Python Labs': 'red', 'Economics': 'purple', 'Statistics Labs': 'Magenta'}

# Plotting
plt.figure(figsize=(10, 6))
for subject in df['Subject'].unique():
    subset = df[df['Subject'] == subject]
    plt.scatter(subset['Student_ID'], subset['Normalized_Score'],
                color=colors.get(subject, 'black'), label=subject)

plt.title('Normalized Scores by Subject')
plt.xlabel('Student ID')
plt.ylabel('Normalized Score (0 to 1)')
plt.legend(title='Subject')
plt.show()
```

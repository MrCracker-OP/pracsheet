Explanation:

The simple_logger decorator defines a wrapper function that logs the name of the original function using func.__name__.

The decorator is applied to the greet function using the @simple_logger syntax.

When greet() is called, the decorator logs the function's name before executing the greeting.

```python
# Define the decorator to log the function name
def simple_logger(func):
    def wrapper():
        print(f"Executing function: {func.__name__}")  # Log the function name
        func()  # Execute the original function
    return wrapper

# Apply the decorator to a function that prints a greeting message
@simple_logger
def greet():
    print("Hello, welcome!")

# Call the decorated function and display the logs
greet()
```

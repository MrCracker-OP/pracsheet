Split() : splits the string into words.
Join() : joins the words without spaces.
lower() : converts the string into a lowercase.

```python
# Input string
input_string = "A man a plan a canal Panama"

# Remove spaces and convert to lowercase
cleaned_string = ''.join(input_string.split()).lower()

# Check if the cleaned string is equal to its reverse
is_palindrome = cleaned_string == cleaned_string[::-1]

# Print the result
if is_palindrome:
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome.')
```

Accept input string from the user

```python
# Accept input string from the user
input_string = input("Enter a string to check if it is a palindrome: ")

# Remove spaces and convert to lowercase
cleaned_string = ''.join(input_string.split()).lower()

# Check if the cleaned string is equal to its reverse
is_palindrome = cleaned_string == cleaned_string[::-1]

# Print the result
if is_palindrome:
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome.')
```

Expt 14: Statistical Analysis with NumPy

Explanation:
Mean, Median, and Standard Deviation:

np.mean(temperatures) calculates the average temperature.
np.median(temperatures) finds the middle value when the temperatures are sorted.
np.std(temperatures) computes the standard deviation, which measures the amount of variation in the temperature values.
Days with Temperature Above the Mean:

np.where(temperatures > mean_temp) returns the indices of the days where the temperature is above the mean.
The indices are adjusted by adding 1 to match the day number (assuming the array index starts from 0).
Plotting a Histogram:

plt.hist(temperatures, bins=5, color='blue', edgecolor='black') creates a histogram with 5 bins.
The histogram displays the distribution of temperature frequencies.
This script will give you the mean, median, standard deviation, identify days with temperatures above the mean, and display a histogram of the temperatures.

```python
import numpy as np
import matplotlib.pyplot as plt

# Temperature dataset
temperatures = np.array([22, 25, 24, 30, 28, 26, 27, 29, 31, 32])

# 1. Calculate mean, median, and standard deviation
mean_temp = np.mean(temperatures)
median_temp = np.median(temperatures)
std_dev_temp = np.std(temperatures)

print(f"Mean Temperature: {mean_temp}")
print(f"Median Temperature: {median_temp}")
print(f"Standard Deviation: {std_dev_temp}")

# 2. Identify days where the temperature was above the mean
days_above_mean = np.where(temperatures > mean_temp)[0] + 1  # Adding 1 to match day number (index starts from 0)
print(f"Days with temperature above mean: {days_above_mean}")

# 3. Plot a histogram of the temperatures
plt.hist(temperatures, bins=5, color='blue', edgecolor='black')
plt.title('Histogram of Daily Temperatures')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

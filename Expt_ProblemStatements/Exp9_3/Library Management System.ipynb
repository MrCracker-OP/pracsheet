Explanation:

Book Class:

__init__: Initializes a book with title, author, and isbn.

get_details: Returns a formatted string containing the book's details.

__str__: Overrides the string representation for a simpler display of book details.

Library Class:

Contains a list of Book objects.

add_book: Adds a Book object to the library.

remove_book: Removes a book by matching its isbn.

find_book: Finds and returns the details of a book using its isbn.

```python
# Book class definition
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

    def get_details(self):
        return f"{self.title} by {self.author} (ISBN: {self.isbn})"

    def __str__(self):
        return f"{self.title} by {self.author}"

# Library class definition
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f'Book "{book.title}" added to the library.')

    def remove_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print(f'Book "{book.title}" removed from the library.')
                return
        print(f"Book with ISBN {isbn} not found.")

    def find_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                return book.get_details()
        return f"Book with ISBN {isbn} not found."

# Example usage:
if __name__ == "__main__":
    # Create books
    book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565")
    book2 = Book("To Kill a Mockingbird", "Harper Lee", "9780060935467")

    # Create library and add books
    library = Library()
    library.add_book(book1)
    library.add_book(book2)

    # Find a book by ISBN
    print(library.find_book("9780060935467"))  # To Kill a Mockingbird by Harper Lee (ISBN: 9780060935467)

    # Remove a book by ISBN
    library.remove_book("9780743273565")  # Removes The Great Gatsby
    print(library.find_book("9780743273565"))  # Book with ISBN 9780743273565 not found.
```
